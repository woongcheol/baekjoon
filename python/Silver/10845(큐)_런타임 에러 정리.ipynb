```python
# 1차 시도

from sys import stdin

def input_que(message):
    if message[0] == 'pop':
        return que.popleft()
    
    elif message[0] == 'size':
        return len(que)
    
    elif message[0] == 'empty':
        if que:
            return 0
        else:
            return 1
        
    elif message[0] == 'front':
        if not que:
            return -1
        else:
            return que[0]
        
    elif message[0] == 'back':
        if not que:
            return -1
        else:
            return que[-1]
        
    elif message[1]:
        que.append(message[1])
        
N = int(stdin.readline())

que = []

for _ in range(N):
    message = stdin.readline().split()
    print(input_que(message))
```

- 런타임 에러는 내부적으로 구현이 잘못되었을 때 발생한다.
    - 배열에 할당된 크기를 넘어서 접근했을 때
    - 전역 배열의 크기가 메모리 제한을 초과할 때
    - 지역 배열의 크기가 스택 크기 제한을 넘어갈 때
    - 0으로 나눌 떄
    - 라이브러리에서 예외를 발생시켰을 때
    - 재귀 호출이 너무 깊어질 때
    - 이미 해제된 메모리를 또 참조할 때
    - 프로그램(main 함수)이 0이 아닌 수를 반환했을 때
    - C/C++에서 반환형이 void가 아닌 main이 아닌 함수에서 아무런 값을 반환하지 않았을 때
