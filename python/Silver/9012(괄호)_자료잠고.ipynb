```python
# 1차 시도

T = int(input())
left = '('
right = ')'

# VPS 검사
def find_vps(ps):
    while ps:
        for i in range(len(ps)-1):
            if ps[i] == left and ps[i+1] == right:
                del ps[i]
                del ps[i]
                break
        if not ps:
            break
        elif ps.count(left) == 0 or ps.count(right) == 0:
            break

    if not ps:
        print('YES')

    else:
        print('NO')

# 테스트 케이스만큼 반복 실행
for i in range(T):
    ps = list(input())
    
    # vps 찾기
    find_vps(ps)
```
