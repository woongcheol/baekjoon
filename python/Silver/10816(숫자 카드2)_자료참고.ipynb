```python
# 1차 시도
from sys import stdin
_ = stdin.readline()
N = sorted(map(int,stdin.readline().split()))
_ = stdin.readline()
M = map(int,stdin.readline().split())

def binary(n, N, start, end):
    if start > end:
        return 0
    m = (start+end)//2
    if n == N[m]:
        return N[start:end+1].count(n)
    elif n < N[m]:
        return binary(n, N, start, m-1)
    else:
        return binary(n, N, m+1, end)

n_dic = {}
for n in N:
    start = 0
    end = len(N) - 1
    if n not in n_dic:
        n_dic[n] = binary(n, N, start, end)

print(' '.join(str(n_dic[x]) if x in n_dic else '0' for x in M ))
```

```python
# 2차 시도

N = int(input())
card_number = list(map(int, input().split()))

M = int(input())
find_card = list(map(int, input().split()))

find_list = []

for idx in range(M):
    find_list.append(card_number.count(find_card[idx]))

print(find_list)
```

### 2차 시도
- for 문에 O(n)
    - append에 O(1) -> for문으로 O(n)
    - count에 O(n) -> for문으로 O(n^2)
- 따라서, O(n^2)으로 인해 시간초과 발생

```python
# 3차 시도

N = int(input())
card_number = list(map(int, input().split()))

M = int(input())
find_card = list(map(int, input().split()))

find_list = []

for num in find_card:
    cnt = 0
    idx = 0
    while idx < len(card_number): 
        if card_number[idx] == num:
            cnt += 1
            del card_number[idx]
            continue
        idx += 1
    if cnt > 0:
        find_list.append(cnt)
    else:
        find_list.append(0)

print(find_list)
```

### 3차 시도
- 기존 for문에서 값을 찾을 때마다 삭제하면서 진행
- 그럼에도 시간초과 발생
