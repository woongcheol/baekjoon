```python
# 1차 시도

# 2차 평면 점 N개 입력
point_number = int(input())

# 점 리스트 생성
point_list = []

# 좌표 입력
for i in range(point_number):
    x_position, y_position = map(int, input().split())
    point_list.append((x_position, y_position))

# 정렬 - 우선 순위 1(x 증가), 우선 순위 2(y 증가)
point_list.sort()
                    
                    
# 출력
for j in range(point_number):
    print(point_list[j][0], point_list[j][1])
```
