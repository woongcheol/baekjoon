```python
# 1차 시도
def input_stack(message):
    if message[0] == 'pop':
        if not stack:
            print(-1)
        else:
            print(stack.pop())
    
    elif message[0] == 'size':
        print(len(stack))

    elif message[0] == 'empty':
        if not stack:
            print(1)
        else:
            print(0)

    elif message[0] == 'top':
        if not stack:
            print(-1)
        else:
            print(stack[-1])

    elif message[1]:
        stack.append(message[1])

N = int(input())
stack = []

for _ in range(N):
    input_stack(input().split())
```

- 시간 초과가 되어 파이썬 내장 모듈인 sys의 stdin을 사용해보았다
    - 단순 입력일 경우 input()을 사용하지만, for 문 등 반복문에 입력 시 효율이 좋다.

```python
# 표준 입력 import
from sys import stdin

# 메세지 입력 함수 구현
def input_stack(message):
    if message[0] == 'pop':
        if not stack:
            print(-1)
        else:
            print(stack.pop())
    
    elif message[0] == 'size':
        print(len(stack))

    elif message[0] == 'empty':
        if not stack:
            print(1)
        else:
            print(0)

    elif message[0] == 'top':
        if not stack:
            print(-1)
        else:
            print(stack[-1])

    elif message[1]:
        stack.append(message[1])

N = int(input())
stack = []

for _ in range(N):
    input_stack(stdin.readline().split())
```
