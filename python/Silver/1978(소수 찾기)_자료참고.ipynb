```python
import sys
# 1차 시도

# 숫자 N 입력
N = int(input())
number = map(int, input().split())

# 함수 생성 - 소수 찾기
def prime(num):
    if num == 1:
        return False
    elif num == 2:
        return True
    for i in range(2, num):
        if num%i == 0:
            return False
    return True
        
# 함수 출력
count = 0
for num in number:
    if prime(num):
        count += 1

print(count)
```
