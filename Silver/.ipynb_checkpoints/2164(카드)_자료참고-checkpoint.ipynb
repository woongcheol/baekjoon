```python
# 1차 시도

# 카드 번호 지정
N = int(input())

# 카드 생성 
card = [num for num in range(1, N+1)]

# 함수 생성
def card_remain(card):
    
    # 카드 1장이 남을 때까지 반복
    while len(card) != 1:
    
        # 1차 : 카드 제거
        card.pop(0)

        # 2차 : 카드 밑장으로 옮기기
        card.append(card.pop(0))
    
    return card[0]
        
print(card_remain(card))
```

```python
# 2차 시도

# 카드 번호 지정
N = int(input())

# 카드 생성 - 1 제외
card = [num for num in range(2, N+1)]

# 함수 생성
def card_remain(card):
    
    # 카드 1장이 남을 때까지 반복
    while len(card) != 1:
    
        # 2차 : 카드 밑장으로 옮기기 - 2부터 시작
        card.append(card.pop(0))
    
        # 1차 : 카드 제거
        card.pop(0)
    
    return card[0]
        
print(card_remain(card))
```

- 리스트 컴프리헨션 시작을 2로 지정하였음

```python
# 3차 시도 - pypy

# 카드 번호 지정
N = int(input())

# 카드 생성 - 1 제외
card = [num for num in range(2, N+1)]

# 함수 생성
def card_remain(card):
    
    # 카드 1장이 남을 때까지 반복
    while len(card) != 1:
    
        # 2차 : 카드 밑장으로 옮기기 - 2부터 시작
        card.append(card.pop(0))
    
        # 1차 : 카드 제거
        card.pop(0)
    
    return card[0]
        
print(card_remain(card))
```

```python
from collections import deque

N = int(input())
card = deque([num for num in range(1, N+1)])

def card_remain(card):
    while len(card) > 1:
        card.popleft()
        move_num = card.popleft()
        card.append(move_num)

    return card.pop()

print(card_remain(card))
```

- 자료 참고
    - 1번 풀이 : 표준 라이브러리 collections 사용(deque)
    - 2번 풀이(생략) : 규칙 찾기
